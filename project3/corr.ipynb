```python
orig_data = pd.read_excel(io="Project 3 Spreadsheet.xlsx", sheet_name="Final Model", nrows=6)

data = orig_data.copy()
data.set_index(data["City"].values, drop=True, inplace=True)
data.drop("City", inplace=True, axis=1)
data
```

```python
data2 = orig_data.drop("City", axis=1)
data_norm = data2.apply(lambda x: (x - x.mean())/x.std() )
data_norm
```

```python
# unnormalzed
corr = data2.corr()
matrix = np.triu(corr)
sns.heatmap(corr, mask=matrix, annot=True)
```
