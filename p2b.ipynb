```python
def values_over_time(stocks, inflows, outflows):
    times = [10, 25, 50, 100]
    values = {}

    for t in times:
        stock = a.stocks.iloc[t]
        flux = a.inflows.iloc[t] - a.outflows.iloc[t] 
        values[t] = {}
        values[t]["stock"] = stock
        values[t]["flux"] = flux

    return values 
```

```python
a = Model()
a.pollution_lim = True
a.sensitivity = True
senitivity_data = {}
for k,v in a.constants.items():
    senitivity_data[k] = {}
    for delta in [0.5, 1, 1.5]:
        a.constants[k] = v * delta
        a.run_sim(time_steps=101)
        senitivity_data[k][delta] = values_over_time(a.stocks, a.inflows, a.outflows)
        # print(k, senitivity_data[k][100]["stock"]["buildings"])
        # reset the value of the constant after the simulation 
        a.constants[k] = v

# a.constants
```
