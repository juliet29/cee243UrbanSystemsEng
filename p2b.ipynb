## Create Sensitivity Data

```python
def values_over_time(stocks, inflows, outflows):
    times = [10, 25, 50, 100]
    values = {}

    for t in times:
        stock = a.stocks.iloc[t]
        flux = a.inflows.iloc[t] - a.outflows.iloc[t] 
        values[t] = {}
        values[t]["stock"] = stock
        values[t]["flux"] = flux

    return values 
```
